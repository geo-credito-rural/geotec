# <span style="color:#336699">Ambientes de Computação Interativa</span>

Atualmente, o ambiente de computação interativa denominado [Jupyter](https://jupyter.org) tem recebido grande aceitação, tanto pela comunidade acadêmica quanto da indústria, para realização de pesquisas e trabalhos que envolvam o emprego de métodos computacionais.

### <span style="color:#336699">Células Markdown</span>

## Título Nível 2

### Título Nível 3

#### Título Nível 4

### <span style="color:#336699">Células de Código Python</span>
<hr style="border:0.25px solid #0077b9;">

```python
#
# Esta célula é do tipo "código",
# logo você pode escrever qualquer código Python válido!
#
# Para executar essa célula, tecle SHIFT+ENTER
#

lista_quadrados = [ x**2 for x in range( 1, 11 ) ]

soma_quadrados = sum( lista_quadrados )

print( "Lista =", lista_quadrados )

print( "Soma:", soma_quadrados )
```

### <span style="color:#336699">Debugger</span>
<hr style="border:0.25px solid #0077b9;">

Um *debugger* (ou *depurador*) é uma ferramenta essencial de desenvolvimento de código pois permite executar um programa passo a passo, pausá-lo em pontos específicos (*breakpoints*) e inspecionar o estado desse programa (variáveis e memória) em tempo de execução.

```python
print("Olá Mundo!")

soma = 0

for i in range(0, 10):
    print(f"Iteração: {i}")

    soma = soma + i

print(f"A soma final é: {soma}.")
```
